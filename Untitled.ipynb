```python
from random import shuffle
```

```python
text = open("data/combilex_rpc.dct")
data = []
for term in text.read().split('\n'):
    data.append(term)
```

```python
shuffle(data)
```

```python
N=len(data)
```

```python
graphemes = data[:int(N/2)]
phonemes = data[int(N/2):]
```

```python
graphemeFile = open('graphemes.txt','w')
for g in graphemes:
    term = g.split(' ')[0]
    graphemeFile.write(term)
    graphemeFile.write('\n')
graphemeFile.close()
```

```python
phonemeFile = open('phonemes.txt','w')
for p in phonemes:
    phones = ' '.join(p.split(' ')[1:])
    phonemeFile.write(phones)
    phonemeFile.write('\n')
phonemeFile.close()
```
